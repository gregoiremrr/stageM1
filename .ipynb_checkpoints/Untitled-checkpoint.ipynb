```python
import math
import numpy as np
import matplotlib.pyplot as plt

zz = 1000
dt = 1/zz
T = 150

# Paramtres
aR = 5
aS = 10
b = 1
MaxDose = 2
Ncrit = 400
Ncrifinf = Ncrit * 0.3
death_rate = 0.003

# Conditions initiales
n0 = 100
r0 = 30
s0 = n0 - r0

# Contrle
def g(n,r,s): # classic
    if n >= Ncrifinf:
        return 1
    return 0

def mtd(): # MTD
    return 1

def no_therapy(): # no_therapy
    return 0

# Fonctions du systme d'quations diffrentielles
def gr(n, r):
    return (aR * n**(-1/3) - b) * r

def gs(n, c, s):
    return (aS * n**(-1/3) - b) * (1 - MaxDose*c) * s

# Mthode d'Euler
def edo(n0, r0, s0, controle):
    resN = []
    resR = []
    resS = []
    n = n0
    r = r0
    s = s0
    for i in range(zz*T):
        dr = gr(n, r)
        if controle == "classic":
            ds = gs(n, g(n,r,s), s)
        elif controle == "MTD":
            ds = gs(n, mtd(), s)
        elif controle == "no_therapy":
            ds = gs(n, no_therapy(), s)
        r = r + dt*dr
        s = s + dt*ds
        n = r + s
        resN.append(n)
        resR.append(r)
        resS.append(s)
        
        if i % zz == 0:
            print("Au temps t = ", i/zz, ", N = ", n, ", R = ", r, ", S = ", s, "N/K = ", n/((a/b)**3))
        
    return resN, resR, resS

x = np.arange(0,T,dt)

# Adaptative treatment
evN, evR, evS = edo(n0, r0, s0, "classic")
plt.plot(x, evN, "r")
plt.plot(x, evR, "b")
plt.plot(x, evS, "g")
plt.show()

# Maximal tolerated dose
evN_MTD, evR_MTD, evS_MTD = edo(n0, r0, s0, "MTD")
plt.plot(x, evN_MTD, "r")
plt.plot(x, evR_MTD, "b")
plt.plot(x, evS_MTD, "g")
plt.show()

# No treatment
evN_NT, evR_NT, evS_NT = edo(n0, r0, s0, "no_therapy")
plt.plot(x, evN_NT, "r")
plt.plot(x, evR_NT, "b")
plt.plot(x, evS_NT, "g")
plt.show()

# Total tumour burden
plt.plot(x, np.array(evN)/((aR/b)**3), "r")
plt.plot(x, np.array(evN_MTD)/((aR/b)**3), "b")
plt.plot(x, np.array(evN_NT)/((aR/b)**3), "g")
plt.show()

# Probability of being alive at time t
def death_curve(evN):
    res = []
    integrale = 0
    for i in range(zz*T):
        integrale += dt*evN[i]
        prob = math.exp(- death_rate * integrale)
        res.append(prob)
    return res

plt.plot(x, death_curve(evN), "g")
plt.plot(x, death_curve(evN_MTD), "b")
plt.plot(x, death_curve(evN_NT), "r")
plt.show()
```
